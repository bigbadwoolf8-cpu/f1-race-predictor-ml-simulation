```python
# -------------------------------
# IMPORTS
# -------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from data_loader import load_event
from model import prepare_data, train_ranking_model
from simulator import simulate_race, summarize

# -------------------------------
# LOAD DATA
# -------------------------------
df = load_event(2023, 'Monza')
print(df.head())

# -------------------------------
# PREPROCESS
# -------------------------------
df = prepare_data(df)
print(df.head())

# -------------------------------
# TRAIN MODEL
# -------------------------------
model, features = train_ranking_model(df)

# -------------------------------
# RUN SIMULATION
# -------------------------------
results = simulate_race(df, model, features, sims=500)
summary = summarize(results)

print(summary)

# -------------------------------
# WIN PROBABILITY PLOT
# -------------------------------
plt.figure()
plt.bar(summary['driver'], summary['win_prob'])
plt.xticks(rotation=90)
plt.title("Win Probability")
plt.xlabel("Driver")
plt.ylabel("Probability")
plt.show()

# -------------------------------
# PODIUM PROBABILITY PLOT
# -------------------------------
plt.figure()
plt.bar(summary['driver'], summary['podium_prob'])
plt.xticks(rotation=90)
plt.title("Podium Probability")
plt.xlabel("Driver")
plt.ylabel("Probability")
plt.show()

# -------------------------------
# DISTRIBUTION PLOT (TOP DRIVER)
# -------------------------------
driver = summary.iloc[0]['driver']
positions = results[driver]

plt.figure()
plt.hist(positions, bins=20)
plt.title(f"Finish Position Distribution: {driver}")
plt.xlabel("Position")
plt.ylabel("Frequency")
plt.show()
```
